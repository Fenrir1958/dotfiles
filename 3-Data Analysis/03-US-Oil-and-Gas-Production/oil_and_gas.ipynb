## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df['Month'].dt.year,0).sum()

yearly_gas_df
```

### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()
print(result.check())
```

```python
plot = yearly_gas_df.filter(items=['U.S.']).plot(kind="bar",title="total production in the U.S.")
```

```python
filtered_yearly_gas_df = yearly_gas_df[np.logical_and(yearly_gas_df.index>2008,yearly_gas_df.index<2018)]
```

```python
filtered_yearly_gas_df
```

```python
plot = filtered_yearly_gas_df.filter(items=['U.S.']).plot(kind="bar",title="total production in the U.S.")
```

## Comparing with Crude Oil Production

```python
file = "data/U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",",converters={'Month':pd.to_datetime})
oil_df.head()
```

```python
oil_df.info()
```

---

## Yearly Oil production

```python
yearly_oil_df = oil_df.groupby(oil_df['Month'].dt.year,0).sum()

yearly_oil_df
```

```python
yearly_oil_df.columns
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
```

```python
yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar');
```

```python
filtered_yearly_oil_df = yearly_oil_df[np.logical_and(yearly_oil_df.index>2008,yearly_oil_df.index<2018)]
```

### Check your code

```python
from nbresult import ChallengeResult
result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()
print(result.check())
```

## Merging Oil and Gas Production

```python
total_gas = pd.DataFrame(filtered_yearly_gas_df['U.S.'])
total_oil = pd.DataFrame(filtered_yearly_oil_df['U.S. Crude Oil'])
```

```python
total_gas.rename({'U.S.':'Gas'},axis=1, inplace=True)

total_gas
```

```python
total_oil.rename({'U.S. Crude Oil':'Crude Oil'}, axis=1, inplace=True)

total_oil
```

```python
merged_df = pd.concat([total_gas,total_oil],axis=1)
```

```python
merged_df
```

```python
plot = merged_df.plot(kind="bar")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
