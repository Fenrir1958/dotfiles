## Merge Practice

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

### Inner Merge Practice


```python
inner_merged_df = a_df.merge(b_df)

inner_merged_df
```

#### Check your code

## Olympic Sports and Medals, 1896-2014

```python
file = "dictionary.csv"
countries_df = pd.read_csv(file, delimiter=",")

countries_df
```

```python
file = "summer.csv"
summer_df = pd.read_csv(file, delimiter=",")

summer_df
```

```python
file = "winter.csv"
winter_df = pd.read_csv(file, delimiter=",")

winter_df
```

### Combining The Data

```python
summer_df.rename({'Country': 'Code'}, axis=1, inplace=True)

summer_df
```

```python
winter_df.rename({'Country': 'Code'}, axis=1, inplace=True)

winter_df
```

```python
summer_countries_df = countries_df.merge(summer_df)

summer_countries_df['Season'] = "Summer"

summer_countries_df
```

```python
winter_countries_df = countries_df.merge(winter_df)

winter_countries_df['Season'] = "Winter"

winter_countries_df
```

```python
all_df = pd.concat([summer_countries_df , winter_countries_df ], axis="index", sort=False)

all_df
```

#### Check your code

```python
from nbresult import ChallengeResult
result = ChallengeResult('all_df',
    all_df_shape=all_df.shape,
    all_df_columns=set(all_df.columns)
)
result.write()
print(result.check())
```

### Top Countries Analysis

```python
top_10_df = all_df[all_df['Year']>=1984]

top_10_df = top_10_df.groupby(top_10_df['Country'])

top_10_df = pd.DataFrame(top_10_df['Medal'].count())

top_10_df.rename({'Medal': 'Medal Count'}, axis=1, inplace=True)

top_10_df = top_10_df.sort_values(by=['Medal Count'],ascending=False)

top_10_df = top_10_df[:10]

top_10_df
```

```python
top_10_df.plot(kind="bar");
```

#### Check your code

```python
from nbresult import ChallengeResult
result = ChallengeResult('olympic_games',
    top_country_1=top_10_df.iloc[0]['Medal Count'],
    top_country_10=top_10_df.iloc[9]['Medal Count']
)
result.write()
print(result.check())
```

```python
top_10_season_df = all_df[all_df['Year']>=1984]

top_10_season_df = top_10_season_df.groupby([top_10_season_df['Country'],top_10_season_df['Season']])

top_10_season_df = pd.DataFrame(top_10_season_df['Medal'].count())

top_10_season_df = top_10_season_df['Medal'].unstack(level=-1)

top_10_season_df = top_10_season_df.sort_values(by=['Summer','Winter'],ascending=False)

top_10_season_df = top_10_season_df[:10]

top_10_season_df['Total'] = top_10_season_df.sum(axis=1)

top_10_season_df
```

```python
top_10_season_df[['Summer', 'Winter']].plot(kind="bar");
```

#### Check your code

```python
from nbresult import ChallengeResult
result = ChallengeResult('olympic_games_season',
    top_country_season_shape=top_10_season_df.shape,
    top_country_1_summer=top_10_season_df.iloc[0]['Summer'],
    top_country_10_winter=top_10_season_df.iloc[9]['Winter']
)
result.write()
print(result.check())
```
